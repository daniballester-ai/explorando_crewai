```python
import os
from crewai import Agent, Task, Crew, Process, LLM
from google.colab import userdata
from langchain_google_genai import ChatGoogleGenerativeAI
import warnings
warnings.filterwarnings('ignore')

os.environ['GEMINI_API_KEY'] = userdata.get('GEMINI_API_KEY')
```

```python
# Inicializando um LLM
llm = LLM(
    model="gemini/gemini-2.0-flash",
    temperature=1,
)

# criação de um agente
agente = Agent(
    role='Criador de rascunhos',
    goal='Escrever ideias iniciais para textos criativos',
    backstory='Um escritor ágil especializado em criar listas e ideias rapidamente',
    llm=llm # Pass the Gemini LLM to the agent
)

# Tarefa
tarefa_gerar_ideias = Task(
    description='Criar 3 ideias curtas para posts sobre os benefícios de estudar o CrewAI',
    expected_output='3 bullet points, comece com um emoji, cada um com uma ideia curta e criativa',
    agent=agente,
    output_file='ideias_crewai.md'
)

# Crew
equipe = Crew(
    agents=[agente],
    tasks=[tarefa_gerar_ideias],
    process=Process.sequential,
    manager_llm=llm # Pass the Gemini LLM to the crew manager
)

# Kickoff
resultado = equipe.kickoff()
print(resultado)
```
